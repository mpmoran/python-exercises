```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
```

```python
def range_tenths(start: int, stop: int) -> list:
    """
    inclusive of stop
    """
    tenths = []
    for n in range(start, stop):
        for i in range(10):
            tenths.append(round(n + i / 10, 1))
        if n == stop - 1:
            tenths.append(stop)
    return tenths

range_tenths(-4, 4)
```

2. Use matplotlib to plot the following equation: y = x^2 - x + 2. Add an anotation for the point (0, 0), the origin.

```python
x = range(-10, 11)
y = [n**2 - n + 2 for n in x]
plt.plot(x, y)
plt.title("$y=x^2-x+2$")
plt.xticks(x[::2])
# plt.annotate('Origin', xy=(0, 0), xytext=(-1, 25), fontsize=14, color='black', arrowprops={'facecolor': 'black'})
plt.text(0, 6, '(0, 0)', fontsize=12)
plt.show()
```

3. Create and label 4 separate plots for the following equations (choose a range for x that makes sense):
    * y = sqrt(x)
    * y = x^3
    * y = tan(x)
    * y = 2^x
You can use functions from the math module to help implement some of the equations above.

```python
def square_root_x():
    # y = sqrt(x)
    label = "$y=\sqrt{x}$"
    x1 = range_tenths(0, 10)
    y1 = [math.sqrt(n) for n in x1]
    plt.plot(x1, y1, label=label)
    plt.title(label)
    
square_root_x()
plt.show()
```

```python
def x_cubed():
    # y = x^3
    label = "$y=x^3$"
    x2 = range_tenths(-3, 3)
    y2 = [n**3 for n in x2]
    plt.plot(x2, y2, label=label)
    plt.title(label)
    plt.xticks(range(-3, 4))
    
x_cubed()
plt.show()
```

```python
def tan_x():
    # y = tan(x)
    label = "$y=\\tan(x)$"
    x3 = range_tenths(-4, 4)
    y3 = [math.tan(n) for n in x3]
    plt.plot(x3, y3, label=label)
    plt.title(label)
    
tan_x()
plt.show()
```

```python
def two_raised_to_x():
    # y = 2^x
    label = "$y=2^x$"
    x4 = range_tenths(-3, 5)
    y4 = [2**n for n in x4]
    plt.plot(x4, y4, label=label)
    plt.title(label)

two_raised_to_x()
plt.show()
```

4. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(15, 10))
plt.suptitle("Four Subplots")

# first subplot
plt.subplot(221)
square_root_x()

# second subplot
plt.subplot(222)
x_cubed()

# third subplot
plt.subplot(223)
tan_x()

# fourth subplot
plt.subplot(224)
two_raised_to_x()

plt.show()
```

5. Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
plt.figure(figsize=(15, 10))
square_root_x()
x_cubed()
tan_x()
two_raised_to_x()
plt.title("Four Equations in One Figure")
plt.xlim(-4, 11)
plt.xticks(range(-4, 11))
plt.legend(fontsize="x-large")
plt.show()
```
